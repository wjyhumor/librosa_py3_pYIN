```python
import numpy as np
from pyin import pyin
import matplotlib.pyplot as plt
```

```python
file = '../audio-archive/chinesewoman1/yin4-cw1.wav'

pitch, rms = pyin(file, inputSampleRate=None, lowAmp=0.01)
print(len(pitch), len(rms))
rms = rms[:len(pitch)]
pitch[pitch<0] = float('nan')
pitch[rms/np.mean(rms)<1.3] = float('nan')
pitch = (pitch-np.nanmean(pitch))/(0.03*np.nanstd(pitch))
pitch = pitch + 300.0

plt.figure(1)
plt.subplot(211)
plt.plot(rms)
axes = plt.gca()
axes.set_xlim([0,len(rms)])
plt.subplot(212)
plt.plot(pitch)
axes = plt.gca()
axes.set_xlim([0, len(pitch)])
axes.set_ylim([100,500])

plt.figure(2)
plt.plot(rms/np.min(rms))
axes = plt.gca()
axes.set_xlim([0,len(rms)])

plt.figure(3)
plt.plot(rms/np.mean(rms))
axes = plt.gca()
axes.set_xlim([0,len(rms)])
```
